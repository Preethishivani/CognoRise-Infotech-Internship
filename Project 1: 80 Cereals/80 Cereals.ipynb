### Data Dictionary

**name :** name of cereal   
**mfr :** Manufacturer of cereal   

                 A = American Home Food Products  
                 G = General Mills   
                 K = Kelloggs  
                 N = Nabisco   
                 P = Post   
                 Q = Quaker Oats    
                 R = Ralston Purina
**type :** type of cereal   

                 C = Cold
                 H = Hot
**calories :** calories per serving     
**protein :** grams of protein   
**fat :** grams of fat   
**sodium :** milligrams of sodium     
**fiber :** grams of dietary fiber    
**carbo :** grams of complex carbohydrates    
**sugars :** grams of sugars    
**potass :** milligrams of potassium   
**vitamins :** vitamins and minerals - 0, 25, or 100, indicating the typical percentage of FDA recommended   
**shelf :** display shelf (1, 2, or 3, counting from the floor)   
**weight :** weight in ounces of one serving   
**cups :** number of cups in one serving    
**rating :** a rating of the cereals (Possibly from Consumer Reports)

### Performing Statistical Analysis on data 

```python
cereal.describe()
```

```python
# Rounding off the data
round(cereal.describe(),2)
```

- **We can see minimum values in carbo,sugars,potass as -1 which can be null values**  
- **Changing -1 to NaN**  

```python
cols_to_change = ['carbo','sugars','potass']
cereal[cols_to_change] = cereal[cols_to_change].replace(-1,np.nan)
```

```python
cereal.head()
```

**Now checking for null values**

```python
cereal.isnull().sum()
```

**Finding mean,median values for carbo,sugars,potass columns**

```python
# carbo column mean and median
print(round(cereal['carbo'].mean(),2))
print(cereal['carbo'].median())
```

```python
# sugars column mean and median
print(round(cereal['sugars'].mean(),2))
print(cereal['sugars'].median())
```

```python
# potass column mean and median
print(round(cereal['potass'].mean(),2))
print(cereal['potass'].median())
```

- **We can see there are three columns which has null values**
- **Replacing null values in carbo,sugars column using mean value of that column**
- **Replacing null values in potass column using medain value of that column**

**Filling null values in carbo,sugars and potass columns**

```python
# carbo column
cereal['carbo'].fillna(cereal['carbo'].mean(),inplace=True)
```

```python
# sugars column
cereal['sugars'].fillna(cereal['sugars'].mean(),inplace=True)
```

```python
#potass column
cereal['potass'].fillna(cereal['potass'].median(),inplace=True)
```

```python
# Again checking for null values
cereal.isnull().sum()
```

**Data is now cleaned and ready to find insights**

## Finding Insights from the cleaned data

```python
cereal.head(2)
```

### 1. Cereals with Highest Nutrients

```python
max_calories_cereal = cereal.loc[cereal['calories'].idxmax()]
max_vitamins_cereal = cereal.loc[cereal['vitamins'].idxmax()]
max_protein_cereal = cereal.loc[cereal['protein'].idxmax()]
max_fat_cereal = cereal.loc[cereal['fat'].idxmax()]
max_sodium_cereal = cereal.loc[cereal['sodium'].idxmax()]
max_fiber_cereal = cereal.loc[cereal['fiber'].idxmax()]
max_carbo_cereal = cereal.loc[cereal['carbo'].idxmax()]
max_sugars_cereal = cereal.loc[cereal['sugars'].idxmax()]
max_potass_cereal = cereal.loc[cereal['potass'].idxmax()]
```

```python
max_values = pd.DataFrame({
    'Nutrient': ['Calories', 'Vitamins','Proteins','Fats','Sodium','Fiber','Carbo','Sugars','Potass'],
    'Cereal': [max_calories_cereal['name'], max_vitamins_cereal['name'],max_protein_cereal['name'],max_fat_cereal['name'],
              max_sodium_cereal['name'],max_fiber_cereal['name'],max_carbo_cereal['name'],max_sugars_cereal['name'],
              max_potass_cereal['name']],
    'Value': [max_calories_cereal['calories'], max_vitamins_cereal['vitamins'],max_protein_cereal['protein'],
              max_fat_cereal['fat'],max_sodium_cereal['sodium'],max_fiber_cereal['fiber'],max_carbo_cereal['carbo'],
              max_sugars_cereal['sugars'],max_potass_cereal['potass']]
})

# Visualize the cereals with the highest nutritional values
plt.figure(figsize=(10, 6))
sns.barplot(x='Nutrient', y='Value', hue='Cereal', data=max_values, palette='Set2')
plt.title('Cereals with the Highest Nutrients')
plt.show()
```

- **Mueslix Crispy Blend cereal has high Calories.**
- **Just RIght Crunchy Nuggets cereal has high vitamins.**
- **Cheerios cereal has high Proteins.**
- **100% Natural Bran cereal has high Fats.**
- **Product 19 cereal has high Sodium Nutrient.**
- **All-Bran with Extra Fiber cereal has high Fiber and pottasium.**
- **Rice Chex cereal has high Carbo content.**
- **Golden Crisp cereal has high Sugars.**

### 2. Cereals with Lowest Nutrients

```python
min_calories_cereal = cereal.loc[cereal['calories'].idxmin()]
min_vitamins_cereal = cereal.loc[cereal['vitamins'].idxmin()]
min_protein_cereal = cereal.loc[cereal['protein'].idxmin()]
min_fat_cereal = cereal.loc[cereal['fat'].idxmin()]
min_sodium_cereal = cereal.loc[cereal['sodium'].idxmin()]
min_fiber_cereal = cereal.loc[cereal['fiber'].idxmin()]
min_carbo_cereal = cereal.loc[cereal['carbo'].idxmin()]
min_sugars_cereal = cereal.loc[cereal['sugars'].idxmin()]
min_potass_cereal = cereal.loc[cereal['potass'].idxmin()]
```

```python
min_values = pd.DataFrame({
    'Nutrient': ['Calories', 'Vitamins','Proteins','Fats','Sodium','Fiber','Carbo','Sugars','Potass'],
    'Cereal': [min_calories_cereal['name'], min_vitamins_cereal['name'],min_protein_cereal['name'],min_fat_cereal['name'],
              min_sodium_cereal['name'],min_fiber_cereal['name'],min_carbo_cereal['name'],min_sugars_cereal['name'],
              min_potass_cereal['name']],
    'Value': [min_calories_cereal['calories'], min_vitamins_cereal['vitamins'],min_protein_cereal['protein'],
              min_fat_cereal['fat'],min_sodium_cereal['sodium'],min_fiber_cereal['fiber'],min_carbo_cereal['carbo'],
              min_sugars_cereal['sugars'],min_potass_cereal['potass']]
})

# Visualize the cereals with the highest nutritional values
plt.figure(figsize=(10, 6))
sns.barplot(x='Nutrient', y='Value', hue='Cereal', data=min_values, palette='Set2')
plt.title('Cereals with the Highest Nutrients')
plt.show()
```

- **All-Bran with Extra Fiber has less calories which is very healthy.**
- **100% Natural Bran has no vitamins.**
- **Cap'n'Crunch has less proteins.**
- **100% Bran has less Carbo.**
- **Puffed Rice has less potass content.**

### 3. Top Rated Cereals

```python
# Top 10 rated cereals
top_rated = cereal.nlargest(10, 'rating')
plt.figure(figsize=(12, 6))
sns.barplot(x='rating', y='name', data=top_rated, palette='viridis')
plt.title('Top 10 Rated Cereals')
plt.show()
```

- **All-Bran with Extra Fiber is the top rated cereal with a rating of 93.7%**

### 4. Manufacturer Comparision

```python
# Average rating by manufacturer
manufacturer_avg = cereal.groupby('mfr')['rating'].mean().sort_values(ascending=False)
print(round(manufacturer_avg,2))
plt.figure(figsize=(10, 6))
sns.barplot(x=manufacturer_avg, y=manufacturer_avg.index, palette='magma')
plt.title('Average Rating by Manufacturer')
plt.show()
```

- **Manufacturer Nabisco has highest ratings of 67.97**
- **Manufacturer American Home Food Products has second highest ratings of 54.85**

### 5. Correlation Analysis

```python
cereal.corr()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(cereal.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Analysis of Nutritional Factors')
plt.show()
```

- **High sugar content generally leads to higher calories**
- **When serving size (in ounces) increases, the calorie content increases significantly. Larger servings generally contain more calories.**
- **Cereals with higher protein content tend to also have higher potassium levels.**
- **If amount of fat in a product increases, then the number of calories tends to increase as well, and vice versa.**
- **If amount of fiber in a product increases,then the amount of potassium tends to increase as well, and vice versa.**
- **If amount of sugars increases, the rating tends to decrease, and vice versa.**
- **If amount of calories increases, the rating tends to decrease, and vice versa.**

### 6. Type Analysis (Cold vs. Hot Cereals)

```python
# Comparison of cold vs. hot cereals
type_comparison = cereal.groupby('type').mean().transpose()
type_comparison.plot(kind='bar', figsize=(12, 8))
plt.title('Nutritional Comparison: Cold vs. Hot Cereals')
plt.xticks(rotation=45)
plt.show()
```

- **Cold type of cereals has high calories,sodium,fiber,sugars,potass,fats,vitamins and shelf life**
- **Hot type of cereals has high proteins,carbohydrates**

### 7. Healthiest Cereals

```python
# Criteria for healthiest cereals: low sugar, low fat, high fiber
healthiest_cereals = cereal[(cereal['sugars'] <= 5) & (cereal['fat'] <= 2) & (cereal['fiber'] >= 5)]
plt.figure(figsize=(12, 6))
sns.barplot(x='fiber', y='name', data=healthiest_cereals, palette='Blues_d')
plt.title('Healthiest Cereals Based on Low Sugar, Low Fat, and High Fiber')
plt.show()
```

- **All-Bran with Extra Fiber cereal is the healthiest cereal with low sugar, low fat and with high fiber**

### 8. Vitamin Content Analysis

```python
# Top cereals by vitamin content
top_vitamins = cereal.nlargest(10, 'vitamins')
plt.figure(figsize=(12, 6))
sns.barplot(x='vitamins', y='name', data=top_vitamins, palette='Blues_d')
plt.title('Top 10 Cereals by Vitamin Content')
plt.show()
```

- **Just RIght Crunchy Nuggets, Just Right Fruit & Nut,Product 19, Total Corn Flakes, Total Raisin Bran, Total Whole Grain cereals has high vitamin levels which is more nutritious**

### 9. Shelf Preference

```python
# Shelf preference analysis
shelf_pref = cereal.groupby('shelf')['rating'].mean().sort_values()
plt.figure(figsize=(10, 6))
sns.barplot(x=shelf_pref.index, y=shelf_pref.values, palette='Pastel2')
plt.title('Average Rating by Shelf Number')
plt.show()
```

- **More number of cereals are in shelf 1 and shelf 3**

### 10. Consumer Preference

```python
# Sweetness (sugar content) vs. Rating
plt.figure(figsize=(12, 6))
sns.scatterplot(x='sugars', y='rating', data=cereal, hue='type', palette='Spectral')
plt.title('Consumer Preference: Sweetness vs. Rating')
plt.show()
```

- **Cold type cerals with almost zero sugar content has high ratings from 60-90+.**
- **Hot type cereals with almost zero sugar content has ratings of 65.**
- **Cold type cereals with high sugar content has ratings in between 30 to 40.**

### 11. Comprehensive Nutritional Score

```python
# Creating a composite score (example: using calories, protein, fiber, and sugars)
cereal['nutritional_score'] = cereal['calories']*0.2 + cereal['protein']*0.3 + cereal['fiber']*0.3 - cereal['sugars']*0.2
top_nutritional_score = cereal.nlargest(10, 'nutritional_score')

plt.figure(figsize=(12, 6))
sns.barplot(x='nutritional_score', y='name', data=top_nutritional_score, palette='coolwarm')
plt.title('Top 10 Cereals by Comprehensive Nutritional Score')
plt.show()
```

- **Mueslix Crispy Blend cereal has a high Nutritional Score.**
- **Muesli Raisins, Dates & Almonds and Muesli Raisins, Peaches & Pecans also has good Nutritional Score.**
