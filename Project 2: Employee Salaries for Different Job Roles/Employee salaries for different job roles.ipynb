### Data Dictionary

**work_year:** The year of employment.    
**experience_level:** The experience level of the employee (e.g., entry-level, mid-level, senior).    
**employment_type:** The type of employment (e.g., full-time, part-time, contract).    
**job_title:** The job title or position of the employee within the company.    
**salary:** The salary amount in the local currency.     
**salary_currency:** The currency in which the salary is denoted.     
**salary_in_usd:** The equivalent salary amount in USD (United States Dollars).    
**employee_residence:** The location of the employee's residence.        
**remote_ratio:** The percentage of remote work allowed for the position.      
**company_location:** The location of the company.    
**company_size:** The size of the company (e.g., small, medium, large).

## Finding Insights from the cleaned data

```python
df.head(2)
```

### 1. Salary Distribution Analysis

**Overall Salary Distribution**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['salary_in_usd'], bins=30, kde=True)
plt.title('Salary Distribution (USD)')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()
```

- **There are more employees whose salary ranges between 90000 USD to 100000 USD**

### 2. Experience Level and Employment Type

**Experience Level Distribution**

```python
plt.figure(figsize=(8, 6))
df['experience_level'].value_counts().plot(kind='bar')
plt.title('Experience Level Distribution')
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.show()
```

- **There are 250+ senior experience level employees**

**Employment Type Distribution**

```python
plt.figure(figsize=(8, 6))
df['employment_type'].value_counts().plot(kind='bar')
plt.title('Employment Type Distribution')
plt.xlabel('Employment Type')
plt.ylabel('Count')
plt.show()
```

- **There are almost 500+ Full time employees**

### 3. Geographical Analysis

**Remote Work Analysis**

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['remote_ratio'], bins=10, kde=True)
plt.title('Remote Work Ratio Distribution')
plt.xlabel('Remote Ratio (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
df['remote_ratio'].value_counts()
```

- **There are so many employees who work in 100% remote ratio**

### 4. Currency Analysis

```python
plt.figure(figsize=(8, 6))
df['salary_currency'].value_counts().plot(kind='bar')
plt.title('Currency Distribution')
plt.xlabel('Currency')
plt.ylabel('Count')
plt.show()
```

- **Most employees receive their salary in USD**

### 5. Correlation Analysis

```python
# Correlation between Salary and Experience Level
experience_salary_corr = df[['salary_in_usd', 'experience_level']].corr()

# Correlation between Remote Ratio and Salary
remote_salary_corr = df[['remote_ratio', 'salary_in_usd']].corr()

# Heatmap of Correlations
plt.figure(figsize=(8, 6))
sns.heatmap(df[['salary_in_usd', 'remote_ratio']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

- **Remote work flexibility has minimal impact on salary levels.**

### 6. Trends Over Time

**Yearly Salary Trends**

```python
yearly_salary = df.groupby('work_year')['salary_in_usd'].median()
plt.figure(figsize=(8, 6))
yearly_salary.plot(marker='o')
plt.title('Median Salary by Year')
plt.xlabel('Year')
plt.ylabel('Median Salary (USD)')
plt.show()
```

- **Salaries of the employees has increased over the years.**

**Experience Level Over Time**

```python
experience_year = df.groupby(['work_year', 'experience_level']).size().unstack().fillna(0)
plt.figure(figsize=(12, 8))
experience_year.plot(marker='o')
plt.title('Experience Level Distribution Over Time')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

- **The proportion of senior-level employees has increased over time.**

### 7. Job Title Popularity

**Most Common Job Titles**

```python
plt.figure(figsize=(8, 6))
df['job_title'].value_counts().head(10).plot(kind='barh')
plt.title('Top 10 Most Common Job Titles')
plt.xlabel('Count')
plt.ylabel('Job Title')
plt.show()
```

- **There are more employees in the job role Data Scientist**

**Job Titles by Employment Type**

```python
job_title_employment = df.groupby(['job_title', 'employment_type']).size().unstack().fillna(0)
plt.figure(figsize=(12, 8))
sns.heatmap(job_title_employment, cmap='YlOrBr', linewidths=0.5)
plt.title('Job Titles by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Job Title')
plt.show()
```

- **There is a high correlation between full time employees and the employees in job roles Machine Learning Engineer, Data Scientist, Data Analyst, Data Architect**
- **Companies prefer more full time employees for specific roles**

### 8. Employee Residence vs. Company Location

**Local vs. International Employment**

```python
df['local_vs_remote'] = df.apply(lambda row: 'Local' if row['employee_residence'] == row['company_location'] else 'Remote', axis=1)
local_remote_count = df['local_vs_remote'].value_counts()
plt.figure(figsize=(8, 6))
local_remote_count.plot(kind='bar')
plt.title('Local vs. Remote Employment')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

### 9. Diversity in Employment Types

```python
# Employment Type by Company Size
employment_company_size = df.groupby(['company_size', 'employment_type']).size().unstack().fillna(0)
plt.figure(figsize=(10, 6))
employment_company_size.plot(kind='bar', stacked=True)
plt.title('Employment Type by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Count')
plt.show()
```

- **There are more Full time employees in all different company sizes.**
- **Medium size company holds more full time employees**

```python
# Experience Level by Employment Type
experience_employment = df.groupby(['experience_level', 'employment_type']).size().unstack().fillna(0)
plt.figure(figsize=(10, 6))
experience_employment.plot(kind='bar', stacked=True)
plt.title('Experience Level by Employment Type')
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.show()
```

- **There are more full time employees in different experience level**
- **There are more senior employees who are working full time**

```python
# Remote Work by Employment Type
remote_employment_type = df.groupby(['remote_ratio', 'employment_type']).size().unstack().fillna(0)
plt.figure(figsize=(10, 6))
remote_employment_type.plot(kind='bar', stacked=True)
plt.title('Remote Work by Employment Type')
plt.xlabel('Remote Ratio (%)')
plt.ylabel('Count')
plt.show()
```

- **Most of the full time employees work in 100% remote ratio**

### 10. Salary Progression by Job Title

**Salary Growth Over Time by Job Title**

```python
top_jobs = df['job_title'].value_counts().index[:10]
plt.figure(figsize=(8, 8))
sns.lineplot(x='work_year', y='salary_in_usd', hue='job_title', data=df[df['job_title'].isin(top_jobs)])
plt.title('Salary Progression by Job Title')
plt.xlabel('Year')
plt.ylabel('Salary (USD)')
plt.show()
```

- **We can see that Data analyst jobs has more salary in 2021 than in 2022**
- **Other job_titles salaries are increased over the time.**

**Career Path Analysis**

```python
# Example: Analyzing for a specific career path like 'Data Scientist' evolving to 'Senior Data Scientist'
career_path = df[df['job_title'].isin(['Data Scientist', 'Senior Data Scientist'])]
plt.figure(figsize=(8, 8))
sns.lineplot(x='work_year', y='salary_in_usd', hue='job_title', data=career_path)
plt.title('Career Path Analysis: Data Scientist to Senior Data Scientist')
plt.xlabel('Year')
plt.ylabel('Salary (USD)')
plt.show()
```

- **As time goes on salray and role of a data scientist increases from Data Scientist to Senior Data Scientist**

**Experience and Salary Change by Job Title**

```python
plt.figure(figsize=(8, 8))
sns.lineplot(x='experience_level', y='salary_in_usd', hue='job_title', data=df[df['job_title'].isin(top_jobs)])
plt.title('Experience Level and Salary Change by Job Title')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.show()
```

- **Salary of a Principal Data Scientist increases as their experience increases**

### 11. Companies Offering the Highest Salaries

```python
top_companies = df.groupby('company_location')['salary_in_usd'].median().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_companies.plot(kind='barh')
plt.title('Top 10 Company locations Offering the Highest Salaries')
plt.xlabel('Median Salary (USD)')
plt.ylabel('Company Location')
plt.show()
```

- **Company Location in Russia is offering high salaries to the employees**
- **Companies located in The United States is offereing salaries ranging between 120000 to 140000**
